## 1. Print Length Times
Write a program that reads a string and prints the first character of the given string on N lines, where N is the length of the given string.


```python
s = input("Enter a string: ").upper()
n = int(input("Enter length: "))
for i in range(1, n+1):
    print(s[0])
```

## 2. Print the Characters
Write a program that reads a string and prints each character of the given string on a new line.


```python
s = input("Enter a string: ")
for i in s:
    print(i)
```

## 3. Square - 2
Write a program that reads a number N and prints a square of N rows and N columns using numbers starting from 1.
 Natural Numbers


```python
n = int(input("Enter a number: "))
for i in range(1, n+1):
    for j in range(1, n+1):
        print(i, end=" ")
    print()
    
```

##  4. Rectangle
Write a program that reads two numbers M and N and prints a rectangle of M rows and N columns using stars (*).

```python
n = int(input("Enter row: "))
m = int(input("Enter column: "))
for i in range(1, n+1):
    for j in range(1, m+1):
        print("*", end="")
    print()
    
```

## 5. Right Angled Triangle
Write a program that reads a number N and prints a right-angled triangle of N rows using stars (*).


```python
n = int(input("Enter a number: "))
for i in range(1, n + 1):
    print('*' * i)
```

## 7. Right Angled Triangle - 3
Write a program that reads a number N and prints a right-angled triangle of N rows using stars (*) and pluses (+).

- The first N-1 rows contain stars (*)
- The Nth row contains pluses (+)


```python
n = int(input("Enter a number: "))
for i in range(1, n + 1):
    if i==n:
        print("+" * n)
    else:
        print("*" * i)
    
```

## 13. Average of N Inputs
Given a number N, write a program that reads N inputs and prints the average of the N inputs.


```python
n = int(input("Enter iteration number: "))

avg = 0

for i in range(n):
    k = int(input("Enter number: "))
    avg += k
avg /= n
print(avg)
```

## 14. Product of N numbers after X
Write a program that reads two numbers X and N and prints the product of N numbers after X.



```python
n = int(input("Enter number: "))
m = int(input("Enter number after n: "))

prod = 1

for i in range(n+1, n+m+1):
    prod *= i
print(prod)
```

## 15. Sum of N Squares
Write a program that reads a number N and prints the sum of squares of N numbers starting from 1.

Example (N = 4):
1² + 2² + 3² + 4² = 30



```python
n = int(input("Enter a number: "))
s = 0
for i in range(1, n + 1):
    s = s + i*i
print(s)
```

## 16. Print Integers - 3
Given an integer N, write a program to print integers from N to 1.
