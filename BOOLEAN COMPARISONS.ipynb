## Problem 1 — Greater Than 70
Problem Statement
- Write a program to check whether the given number is greater than 70.


```python
num = int(input("Enter a number: "))
if num >=70:
    print(True)
else:
    print(False)
```

Problem Statement

```python
num = int(input("Enter a number: "))
num2 = int(input("Enter another number: "))
if num>= num2:
    print("A>=B:",True)
else:
    print(False)
```

## Problem 9 — Consecutive Numbers
Problem Statement
- Write a program to check whether B is exactly one greater than A.


```python
A = int(input("Enter a number: "))
B = int(input("Enter another number: "))
if B == A + 1:
    print("B is one more than A:",True)
else:
    print(False)
```

## Problem 11 — Sum of Digits Check
Problem Statement
- Write a program to check whether the sum of the digits of a two-digit number is greater than 7.


```python
num = int(input())

dig1 = num // 10
dig2 = num % 10

if dig1 + dig2 > 7:
    print(True)
else:
    print(False)
```
