## Problem 1 — Compare Digits
**Problem Statement: Write a program that reads a two-digit number N and checks:**
- If the number N is greater than 25.
- If the first digit of N is greater than the second digit of N.


```python
num = int(input("Enter a number: "))
dig1 = num// 10
dig2 = num%10
print(num >=25)
print(dig1 > dig2)

```

## Problem 2 — Compare First Digit & Last Digit
Problem Statement
- Write a program that reads two three-digit numbers A and B and checks if the first digit of A is less than the last digit of B.


```python
num = int(input("Enter a 3 digit number: "))
num2 = int(input("Enter another 3 digit number: "))
dig1 = num// 100
dig2 = num2%10
if dig1 < dig2:
    print(True)
else:
    print(False)
```

## Problem 3 — Compare Last Three Characters
Problem Statement
- Write a program to check if the last three characters in the two given strings are the same.


```python
st = input("Enter a string: ")
st2 = input("Enter another string: ")
if st[-3:] == st2[-3:]:
    print(True)
else:    
    print(False)
```

## Problem 4 — Percentage of 500
Problem Statement
 - Write a program that reads a percentage P and a number N and checks if P% of 500 is equal to N.


```python
per = int(input("Enter a percentage: "))
n = int(input("Enter a number: "))
p = (per/100)*500
if p ==n:
    print(True)
else:
    print(False)
```

## Problem 5 — Area vs Perimeter of Rectangle
Problem Statement
 - Write a program that checks whether the area of a rectangle is less than or equal to the perimeter.


```python
l = int(input("Enter a length: "))
b = int(input("Enter a breadth: "))
a_rect = l*b
per_rect = 2*(l+b)
if a_rect < per_rect:
    print(True)
else:
    print(False)
```
