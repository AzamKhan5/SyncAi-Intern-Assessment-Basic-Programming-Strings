#**26. Star Repetition – 3**
Problem Statement: **Write a program that reads a string and prints the first and last characters of the given string and prints stars (*) instead of the remaining characters.**


```python
def star():
    n = input("ENTER ANY String: ")
    if n<=2:
        print(n)
    else:
        f = n[0]
        l = n[-1]
        print(f+((len(n)-2)*"*")+l)

star()
```
