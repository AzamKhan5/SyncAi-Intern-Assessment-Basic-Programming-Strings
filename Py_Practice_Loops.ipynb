# Loops, Patterns & Number Programs  

## 1. Print 0, N Times


```python
n = int(input("Enter a number: "))
for i in range(1, n + 1):
    print(0)
```

## 2. Print N Integers

```python
n = int(input("Enter a number: "))
for i in range(1, n + 1):
    print(i)
```

 ## 3. First N Natural Numbers


```python
n = int(input("Enter a number: "))
for i in range(1, n + 1):
    print(i)
```

##  4. Print Integers - 2 (M to N)


```python
n = int(input("Enter a number: ")) 
m = int(input("Enter another number: "))
for i in range(n, m + 1):
    print(i)
```

## 5. Average of N Numbers


```python
n = int(input("Enter number: "))
for i in range(1, n+1):
    s = sum(range(1, i+1))

print(s/i)
```

## 6. Cubes of N Numbers


```python
n = int(input("Enter number: "))
for i in range(1, n+1):
    print(i**3)
```

## 7. Solid Rectangle

```python
n = int(input("Enter Rows: "))
m = int(input("Enter Columns: "))
for i in range(n):
    for j in range(m):
        print("*", end="")
    print()
    
```

## 8. 10 Numbers after N


```python
n = int(input())
for i in range(n+1, n+10):
    print(i)
```

## 9. N Numbers after M


```python
n = int(input("Enter a number: ")) 
m = int(input("Enter another number: "))
for i in range(n+1, m+n+1):
    print(i)
```

## 10. Sum of N Numbers from M


```python
n = int(input("Enter a number: ")) 
m = int(input("Enter another number: "))
for i in range(n, m+n):
    s = sum(range(n, i+1))

print(s)
```

## 11. Sum of N Natural Numbers


```python
n = int(input("Enter number: "))
for i in range(1, n+1):
    s = sum(range(1, i+1))

print(s)
```

## 12. Read N Inputs


```python
n = int(input("Enter iteration number: "))

for i in range(n):
    k = int(input("Enter a number: "))
    print(k)
```

## 13. Product of N Inputs


```python
n = int(input("Enter iteration number: "))

prod = 1

for _ in range(n):
    num = int(input("Enter a number: "))
    prod *= num

print(prod)
```

## 14. Sum of Given Numbers


```python
n = int(input("Enter iteration number: "))
s= 0

for _ in range(n):
    k = int(input("Enter a number: "))
    s = s + k
    
print(s)
```

## 15. Solid Right-Angled Triangle


```python
n = int(input("Enter a number: "))
for i in range(1, n + 1):
    print("*" * i)
```

## 16. Identify Mistake – Print Characters


```python
s = input("Enter a string: ")
for char in s:
    print(char)
```
