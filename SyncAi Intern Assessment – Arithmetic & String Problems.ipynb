## 1. Sum of Two Numbers
**Problem Statement: Write a program to read two integers and print their sum.**


```python
def sum():
    a = int(input("Enter a number: "))
    b = int(input("Enter another number: "))
    return a + b

print(sum())
```

## 3. Area of a Rectangle
**Problem Statement: Write a program to calculate the area of a rectangle.**

```python
def area_rec():
    l = int(input("Enter the length of the rectangle: "))
    w = int(input("Enter the width of the rectangle: "))
    return l * w

print(area_rec())
```

## 4. Perimeter of Rectangle
**Problem Statement: Write a program to calculate the perimeter of a rectangle.**


```python
def perimeter_rec():
    l = int(input("Enter the length of the rectangle: "))
    w = int(input("Enter the width of the rectangle: "))
    return 2 * (l + w)

print(perimeter_rec())
```

## 12. Sum of Digits
**Problem Statement: Write a program to find the sum of the digits of a number.**


```python
def sum_dig():
    num = input("Enter a number: ")
    for d in num:
        return sum(int(d))
    
print(sum_dig())
```

## 13. Indexing

**Problem Statement: Write a program to read a string and an index and print the character at that index.** 


```python
s = input("Enter a string: ")
c= input("Enter a character: ")
for i in s:
    if c in s:
        print("Character found in the index", s.index(c))
        break
```
