## 14. Electricity Bill
Input Explanation
- A single integer units.


```python
units = int(input())

if units <= 50:
    bill = units * 2
elif units <= 150:
    bill = (50 * 2) + (units - 50) * 3
elif units <= 250:
    bill = (50 * 2) + (100 * 3) + (units - 150) * 5
else:
    bill = (50 * 2) + (100 * 3) + (100 * 5) + (units - 250) * 8

bill += bill * 0.20
print(float(bill))
```

## 18. Simple Calculator


```python
O = input("Enter an operator +, -, *, /, %: ")
A = int(input("Enter a number: "))
B = int(input("Enter another number: "))

if O == "+":
    print(A + B)
elif O == "-":
    print(A - B)
elif O == "*":
    print(A * B)
elif O == "%":
    print(A % B)
elif O == "/":
    if B != 0 and A != 0:
        print(A / B)
    else:
        print("Error: Division by zero")
```
